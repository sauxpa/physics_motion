```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
from central_force_app.particle import Particle
```

```python
p = Particle(
    mass=1.0,
    mass_attractive_center=1.0,
    n_steps_ode=int(1e4),
    exponent=-2.0,
)
```

```python
r0 = 1.0
drdt0 = 0.0
theta0 = 0.0
dthetadt0 = 0.9
T = 100.0

t, traj = p.motion(r0, drdt0, theta0, dthetadt0, T)
r, drdt, theta = traj
x = r * np.cos(theta)
y = r * np.sin(theta)
```

```python
with sns.axes_style("darkgrid"):
    fig, ax = plt.subplots(figsize=(11, 11), nrows=1, ncols=1)

    ax.scatter(x[0], y[0], label='Initial position', color='r', s=50)
    ax.plot(x, y, label='Trajectory')
    ax.legend(loc='lower left')
    ax.axis('square')
    
    plt.show()
```
